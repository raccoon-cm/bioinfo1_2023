## Transcript count table 마련하기

제공된 alignment와 annotation 파일들을 이용해서 transcript별 read count를 구해봅니다. 실제 연구에서는 multi-mapping 등을 고려해야 하지만, 여기서는 단순화해서 모두 무시합니다.

```python
!mkdir /casa/chemin/p/bioinfo1_2023/binfo1-work
%cd /casa/chemin/p/bioinfo1_2023/binfo1-work
!cp ../binfo1-datapack1/*.* .
```

```python
!featureCounts -a gencode.gtf -o read-counts.txt *.bam 
```

```python
import pandas as pd
cnts = pd.read_csv('/casa/chemin/p/bioinfo1_2023/binfo1-work/read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```python
cnts
```

```python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```python
cnts.hist(['CLIP-35L33G.bam', 'RNA-control.bam', 'RNA-siLin28a.bam', 'RNA-siLuc.bam', 'RPF-siLin28a.bam', 'RPF-siLuc.bam'], log = True, bins = 100)
```

```python
bam_list = [x for x in cnts.columns if x.split('.')[-1]=='bam']
cnts_filtered = cnts[(cnts[bam_list] >= 100).any(axis = 1)] # read counts 100 이상인 것들만 필터링
```

```python
cnts_filtered
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts_filtered['clip_enrichment']),
           np.log2(cnts_filtered['rden_change']),
           alpha = 0.15,
           s = 2,
           c = 'black'
           )
ax.set_xlim((-6, 6))
ax.set_ylim((-3, 2))
ax.set_xlabel(r'Clip Enrichment (log$_2$ scale)', fontsize = 12)
ax.set_ylabel(r'Ribosome Density Change (log$_2$ scale)', fontsize = 12)
ax.grid(True, zorder=0.01, linestyle = 'dotted')
```

## Protein localization 반영하기

논문 그림에서는 protein localization을 반영해서 색으로 나타냈습니다. 이것을 한 번 구현해 봅시다. 각 gene product의 localization을 따로 구하려면 좀 귀찮기 때문에 미리 모아 두었습니다. UniProt에서 매핑된 데이터를 토대로 아주 명확한 것만 일부 정리한 것입니다. 이것도 마찬가지로 처음엔 에러가 나기도 하니까, 다시 실행해보면 거의 잘 됩니다~

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
mouselocal = pd.read_csv('https://hyeshik.qbio.io/binfo/mouselocalization-20210507.txt', sep='\t')
mouselocal.head()
```

```python
mouselocal
```

```python
cnts_filtered.reset_index()
```

```python
cnts_filtered.reset_index(inplace = True)
cnts_filtered['Geneid_split'] = cnts_filtered.reset_index().apply(lambda x: x['Geneid'].split('.')[0], axis = 1)
cnts_filtered.set_index('Geneid', inplace = True)
```

```python
cnts_filtered
```

```python
cnts_filtered_local = pd.merge(cnts_filtered, mouselocal, left_on='Geneid_split', right_on='gene_id')
```

```python
cnts_filtered_local
```

```python
local = cnts_filtered_local.groupby('type')
for group in local:
    print(group)
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(1, 1, figsize = (5,5))

local = cnts_filtered_local.groupby('type')
colors = {'nucleus': 'dodgerblue', 'integral membrane': 'hotpink', 'cytoplasm': 'limegreen'}



for name, group in local:
    ax.plot(np.log2(group.clip_enrichment),
            np.log2(group.rden_change),
            marker = 'o',
            linestyle = '',
            ms=1,
            c=colors[name],
            label = name,
            alpha = 0.5
            )
ax.legend()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (5,5))

colors = {'nucleus': 'dodgerblue', 'integral membrane': 'hotpink', 'cytoplasm': 'limegreen'}

ax.scatter(np.log2(cnts_filtered_local['clip_enrichment']),
        np.log2(cnts_filtered_local['rden_change']),
        c=[colors[local] for local in cnts_filtered_local['type']],
        s=1)
ax.legend()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (5,5))

colors = {'nucleus': 'dodgerblue', 'integral membrane': 'hotpink', 'cytoplasm': 'limegreen'}
df = cnts_filtered_local.sample(frac=0.5)
ax.scatter(np.log2(df['clip_enrichment']),
        np.log2(df['rden_change']),
        c=[colors[local] for local in df['type']],
        s=1)
ax.set_xlim((-6, 6))
ax.set_ylim((-3, 2))
ax.set_xlabel(r'Clip Enrichment (log$_2$ scale)', fontsize = 12)
ax.set_ylabel(r'Ribosome Density Change (log$_2$ scale)', fontsize = 12)
ax.grid(True, zorder=0.01, linestyle = 'dotted')
```
